# Surface, Toric, and Rotated Surface Code Demo

Small instantiation and visualization example for all three code families.

```python
import sys

# Avoid stale class definitions after local edits.
for mod_name in list(sys.modules):
    if mod_name == 'qldpc_sim' or mod_name.startswith('qldpc_sim.'):
        del sys.modules[mod_name]

from qldpc_sim.qec_code.toric_code import ToricCode
from qldpc_sim.qec_code.surface_code import SurfaceCode
from qldpc_sim.qec_code.rotated_surface_code import RotatedSurfaceCode
```

```python
d = 5

toric = ToricCode.from_distance(d, code_name='td3')
surface = SurfaceCode.from_distance(d, code_name='sd3')
rotated = RotatedSurfaceCode.from_distance(d, code_name='rsd3')


def logical_tag_lines(code):
    lines = []
    for i, lq in enumerate(code.logical_qubits):
        x_tags = [node.tag for node in lq.logical_x.target_nodes]
        z_tags = [node.tag for node in lq.logical_z.target_nodes]
        lines.append(f'LQ{i} X: {x_tags}')
        lines.append(f'LQ{i} Z: {z_tags}')
    return lines


print('Toric   :', f'n={toric.n}, k={toric.k}, d={toric.d}, edges={len(toric.tanner_graph.edges)}')
print('Surface :', f'n={surface.n}, k={surface.k}, d={surface.d}, edges={len(surface.tanner_graph.edges)}')
print('Rotated :', f'n={rotated.n}, k={rotated.k}, d={rotated.d}, edges={len(rotated.tanner_graph.edges)}')
```

```python
fig_toric, ax_toric = toric.tanner_graph.visualize(periodic=True)
lines = logical_tag_lines(toric)
fig_toric.subplots_adjust(right=0.72)
fig_toric.text(0.74, 0.98, '\n'.join(lines), va='top', ha='left', fontsize=9, family='monospace')
print('Toric logical tags:')
for line in lines:
    print('  ', line)
print('Toric visualization rendered.')
```

```python
fig_surface, ax_surface = surface.tanner_graph.visualize(periodic=False)
lines = logical_tag_lines(surface)
fig_surface.subplots_adjust(right=0.72)
fig_surface.text(0.74, 0.98, '\n'.join(lines), va='top', ha='left', fontsize=9, family='monospace')
print('Surface logical tags:')
for line in lines:
    print('  ', line)
print('Surface visualization rendered.')
```

```python
fig_rotated, ax_rotated = rotated.tanner_graph.visualize(periodic=False)
lines = logical_tag_lines(rotated)
fig_rotated.subplots_adjust(right=0.72)
fig_rotated.text(0.74, 0.98, '\n'.join(lines), va='top', ha='left', fontsize=9, family='monospace')
print('Rotated logical tags:')
for line in lines:
    print('  ', line)
print('Rotated surface visualization rendered.')
```
